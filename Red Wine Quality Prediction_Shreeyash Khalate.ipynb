# KNN Classification

```python
# KNN Classification

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

x = mydata.drop(["quality"],axis=1).values
y = mydata["quality"].values


# We will take 20% of our dataset as a test dataset and seed = 24
# and stratify for including every distinct target value possible in dataset in our train set
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,  random_state = 24, stratify = y)

# Creating knn classifier with say 6 neighbors
# (1) Instantiate
knn = KNeighborsClassifier(n_neighbors=21)

# (2) Fit
knn.fit(x_train, y_train)

# (3) Predict
y_pred = knn.predict(x_test)
print("Test set predictions : \n\n {}".format(y_pred))

# Evaluate - Finding out the Accuracy
training_accuracy = knn.score(x_train,y_train)
testing_accuracy = knn.score(x_test,y_test)
print("\nTraining accuracy for our model : ",training_accuracy)
print("Testing accuracy for our model : ",testing_accuracy)
print("\n\nAccuracy of our model : ",testing_accuracy)
```

# Thus accuracy for KNN Classifier model is 56%

```python
# Lets observe accuracy for different values of k
neighbors = np.arange(1,50)
trainaccuracy = np.empty(len(neighbors))
testaccuracy = np.empty(len(neighbors))

#Loop over various k values
for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    trainaccuracy[i] = knn.score(x_train,y_train)
    testaccuracy[i] = knn.score(x_test,y_test)
    
# Generate a plot
plt.title("KNN classifier for varying number of neighbors")
plt.plot(neighbors,testaccuracy, label = "Test Accuracy")
plt.plot(neighbors,trainaccuracy, label = "Train Accuracy")
plt.xlabel("Number of Neighbors")
plt.ylabel("Accuracy")
plt.show()
```

#### Thus, from above graph for larger k values (approx. k=21 to k=31) the accuracy is good as compared to other k values.

# Logistic Regression

```python
# Logistic Regression

# Again reading the dataset for convenience
import pandas as pd
mydata = pd.read_csv("C:/Users/HP/Desktop/Internship Miniproject/4458_8204_bundle_archive/winequality-red.csv")                
mydata.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x = mydata.drop(["quality"],axis=1)
y = mydata["quality"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,  random_state = 33, stratify = y)


# (1) Instantiate
reg = LinearRegression()

# (2) Fit
reg.fit(x_train, y_train)

# (3) Predict
y_pred = reg.predict(x_test)
print("Test set predictions : \n\n",*y_pred, sep = "\n")

```

```python
# Finding out the accuracy
accuracy = reg.score(x_test,y_test)
print("\n\nAccuracy for our model : ",accuracy)
```

# Thus accuracy for Logistic regression model is 44%
